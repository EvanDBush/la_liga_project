```python
import pandas as pd
spain_clean_df = pd.read_csv("results/spain_clean.csv")
team_input = 'FC Barcelona'

team_home_records = spain_clean_df[spain_clean_df['home'] == team_input]
team_visitor_records = spain_clean_df[spain_clean_df['visitor'] == team_input]
team_records_df = [team_home_records, team_visitor_records]
team_records_df = pd.concat(team_records_df).sort_values( by= 'Date')
team_records_df
```
